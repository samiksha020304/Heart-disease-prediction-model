```python
#logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_auc_score
file_path = r"C:\Users\Smita\Downloads\unified_heart_disease.csv"
df = pd.read_csv(file_path)
# Convert gender column to binary
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})
# Select features and target variable
# Assuming 'Heart Disease' is the target column and other columns are features
X = df.drop(columns=['Heart Disease'])
y = df['Heart Disease']
X_encoded = pd.get_dummies(X, drop_first=True)
# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
# Initialize and fit Logistic Regression model
lr = LogisticRegression(max_iter=1000)  # Increase max_iter if needed
model1 = lr.fit(X_train, y_train)
# Make predictions on the test set
predictions1 = model1.predict(X_test)
prediction_probs = model1.predict_proba(X_test)[:, 1]
cm=confusion_matrix(y_test,predictions1)
cm
```

```python
sns.heatmap(cm, annot=True, cmap='BuPu')
```

```python
# Reset print function to its default in case it was overridden
import builtins
print = builtins.print
TP = cm[0][0]
TN = cm[1][1]
FN = cm[1][0]
FP = cm[0][1]

testing_accuracy = (TP + TN) / (TP + TN + FP + FN)
print('Testing Accuracy:', testing_accuracy)
```

```python
accuracy_score(y_test,predictions1)
```

```python
print(classification_report(y_test,predictions1))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
file_path = r"C:\Users\Smita\Downloads\unified_heart_disease.csv"
df = pd.read_csv(file_path)
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})
target_column = 'Heart Disease'
X = df.drop(columns=[target_column])
y = df[target_column]
X_encoded = pd.get_dummies(X)
x_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(X_test)
dtc = DecisionTreeClassifier()
model2 = dtc.fit(x_train_scaled, y_train)
predictions2 = model2.predict(X_test_scaled)
prediction_probs = model2.predict_proba(X_test_scaled)[:, 1]
cm2 = confusion_matrix(y_test, predictions2)

```

```python
#Random Forest
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
file_path = r"C:\Users\Smita\Downloads\unified_heart_disease.csv"
df = pd.read_csv(file_path)
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})
target_column = 'Heart Disease'
X = df.drop(columns=[target_column])
y = df[target_column]
X_encoded = pd.get_dummies(X)
x_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(X_test)
rfc = RandomForestClassifier()
model3 = rfc.fit(x_train_scaled, y_train)
predictions3 = model3.predict(X_test_scaled)
cm3 = confusion_matrix(y_test, predictions3)
print(cm3)
```

```python
accuracy_score(y_test,predictions3)
```

```python
print(classification_report(y_test,predictions3))
```
